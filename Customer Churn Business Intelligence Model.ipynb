# Data Analysis

```python
"""Data Analysis: Here the machine learning technique is selected. 
   
                That's clear, it is classification because in the 
                 target column the value is divided in two groups:
                                        
                                1) 0 for No 
                                2) 1 for Yes. """
```

```python
"""In this model I am using three algorithms:

                1) SVM Algorithm
                2) KNN-Classification Algorithm
                3) Random Forest Algorithm"""
```

# Support Vector Machine(SVM) Algorithm

```python
df0=df[df.Churn==0]
df2=df[df.Churn==1]
```

```python
plt.scatter(df0['MonthlyCharges'], df0['TotalCharges'], color='green', marker='*')
plt.scatter(df2['MonthlyCharges'], df2['TotalCharges'], color='blue', marker='+')
```

```python
# Model Creation ---- SVM(Support Vector Machine) Model
from sklearn.svm import SVC

model=SVC(kernel='rbf')# model is created on training dataset
```

```python
#SVM Model Training
model.fit(X1_train, Y1_train)
```

```python
# SVM Model Testing
# Prediction
Y2_pred=model.predict(X1_test)
Y2_pred
```

```python
from sklearn.metrics import classification_report

print("Classification Report Algorithm 1 is:\n",
      "\n",classification_report(Y1_test, Y2_pred))
```

```python
from sklearn.metrics import accuracy_score
acc1=accuracy_score(Y1_test, Y2_pred)*100
acc1
```

```python
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(Y1_test, Y1_pred)

cm1
```

# KNN-Classification Algorithm

```python
# Model Creation ----KNN
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5) # model is created on training dataset
```

```python
#KNN Model Training
knn.fit(X1_train, Y1_train)
```

```python
#KNN Model Testing
# Prediction
Y2_pred=knn.predict(X1_test)
Y2_pred
```

```python
print("Classification Report Algorithm 2 is:\n",
      "\n",classification_report(Y1_test, Y2_pred))
```

```python
acc2 = accuracy_score(Y1_test, Y2_pred)*100
acc2
```

```python
from sklearn.metrics import confusion_matrix
cm2=confusion_matrix(Y1_test, Y2_pred)

cm2
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm2, annot=True, fmt='.2f')
plt.title("Heatmap on KNN-Classification Model")
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

# RandomForest Algorithm

```python
# Model Creation ---- Random Forest Model
from sklearn.ensemble import RandomForestClassifier as RFC
random = RFC(n_estimators=40)
```

```python
#Random Forest Model Training
random.fit(X1_train, Y1_train)
```

```python
#Random Forest Model Testing
Y3_pred=random.predict(X1_test)
Y3_pred
```

```python
acc3=accuracy_score(Y1_test, Y3_pred)*100
acc3
```

```python
cm3=confusion_matrix(Y1_test, Y3_pred)

cm3
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm3, annot=True, fmt='.2f')
plt.title("Heatmap on Random Forest Model")
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
print("Classification Report Algorithm 3 is:\n",
      "\n",classification_report(Y1_test, Y3_pred))
```
